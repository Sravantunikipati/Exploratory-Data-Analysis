```python
import requests
from bs4 import BeautifulSoup
import re
import numpy as np
brand = []
price = []
fueltype=[]
kilometers=[]
geartype=[]
city_id=[1692,4709,5732,132,5,3686,2920,2130,6356,777,290,2378,2,134,2423,136,769,6105,666,2713,2598,8134,1606,33]
#ahmedabad=requsts.get(r'https://www.cars24.com/buy-used-car?f=make%3A%3D%3Ahonda%3Bmodel%3Ain%3Acity%3AOR%3Amake%3A%3D%3Amaruti%3Bmodel%3Ain%3Abaleno&sort=lhl&serveWarrantyCount=true&gaId=969507114.1699945874&storeCityId=1692&pinId=380001'                                       
for i in  city_id:   
    page = requests.get(f"https://www.cars24.com/buy-used-car?sort=lhl&serveWarrantyCount=true&listingSource=Homepage_Filters&storeCityId={i}")
    soup = BeautifulSoup(page.text)
    for i in soup.find_all('h3', class_='_2lmIw'):
        text = i.text
        regex = re.findall('(\d+\s\w+\s\w+)', text)
        if regex:
            brand.append(regex[0])
        else:
            brand.append(np.nan)

# For Flat type
    for i in soup.find_all('div', class_='_18ToE'):
        text = i.text
        regex = re.findall("(\S\d\S\d+\S\d+)+", text)
        if regex:
            price.append(regex[0])
        else:
            price.append(np.nan)
#For fueltype
    for i in soup.find_all('ul',class_="_13yb6"):
        text = i.text
        regex = re.findall("(Petrol|Diesel|CNG)",text)
        if regex: 
            fueltype.append(regex[0]) 
        else:
            fueltype.append(np.nan)
#For kilometers
    for i in soup.find_all('ul',class_="_13yb6"):
        text = i.text
        regex = re.findall("\d+\S\d+",text)
    
        if regex: 
            kilometers.append(regex[0]) 
        else:
            kilometers.append(np.nan)
#For kilometers
    for i in soup.find_all('ul',class_="_1hOnS"):
        text = i.text
        regex = re.findall("(Manual|Automatic)",text)
    
        if regex: 
            geartype.append(regex[0]) 
        else:
            geartype.append(np.nan)
print(brand)
print(price)
print(fueltype)
print(kilometers)
print(geartype)
```

```python
csv=pd.DataFrame({'Brand':brand,'kilometers_driven':kilometers,'prices':price,'fueltypes':fueltype,'gear_type':geartype})
```
